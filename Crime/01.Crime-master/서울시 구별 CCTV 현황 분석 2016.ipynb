```python
pop2016['여성비율'] = pop2016['여자'] / pop2016['총인구'] * 100
pop2016['남성비율'] = pop2016['남자'] / pop2016['총인구'] * 100
pop2016.head()
```

```python
plt.figure(figsize=(6,6))
plt.scatter(merge2016['총인구'], merge2016['소계'], s=50) #마커의 크기
plt.xlabel('총인구')
plt.ylabel('CCTV')
plt.grid()
plt.show()
```

```python
fp1 = np.polyfit(merge2016['총인구'], merge2016['소계'], 1)
fp1
```

```python
f1 = np.poly1d(fp1)
fx = np.linspace(100000, 700000, 100)
```

```python
plt.figure(figsize=(10,10))
plt.scatter(merge2016['총인구'], merge2016['소계'], s=50)
plt.plot(fx, f1(fx), ls ='dashed', lw = 3, color='g')
plt.xlabel('총인구')
plt.ylabel('CCTV')
plt.grid()
plt.show()
#x축 데이터는 numpy의 linspave로 만들고, y축은 polyld로 만들 수 있다.
```

```python
fp1 = np.polyfit(merge2016['총인구'], merge2016['소계'], 1)

f1 = np.poly1d(fp1)
fx = np.linspace(100000, 700000, 100)

merge2016['오차'] = np.abs(merge2016['소계'] - f1(merge2016['총인구']))

df_sort = merge2016.sort_values(by = '오차', ascending=False)
df_sort.head()
```

```python
plt.figure(figsize=(14,10))
plt.scatter(merge2016['총인구'], merge2016['소계'], c=merge2016['오차'], s=50)
plt.plot(fx, f1(fx), ls='dashed', lw=3, color='g')

for n in range(10):
    plt.text(df_sort['총인구'][n]*1.02, df_sort['소계'][n]*0.98, df_sort.index[n], fontsize=15)
    
plt.xlabel('총인구')
plt.ylabel('인구당비율')

plt.colorbar()
plt.grid()
plt.show()
```

```python
#직선을 기준으로 위에있는 강남구, 관악구, 용산구, 서대문구, 성동구는 서울시 전체 지역의 일반적인 경향보다 CCTV가 많이 서치된 지역이다.
#송파구, 강서구, 중랑구, 도봉구, 강북구는 일반적인 경향보다 CCTV가 적게 설치된 지역이다. 
# 특히, 강남구는 월등히 많은 CCTV가 설치됐지만, 송파구는 인구수에 비해 너무나도 적은 수의 CCTV를 가지고 있다.
```
