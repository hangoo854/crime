```
import pandas as pd
import numpy as np
```

```
data=pd.read_csv("서울특별시_강남구_CCTV_20181008.csv",sep=",",encoding='euc-kr')
```

```
import folium
```

```
Gangnam_map=folium.Map(location=[37.512387,127.047583], zoom_start=13) #location은 대충 강남구 위치로 했음

for n in data.index:
  folium.CircleMarker([data['위도'][n],data['경도'][n]], radius=8, color='#b23776',fill_color='#b23776').add_to(Gangnam_map)

Gangnam_map #강남구 CCTV위치
```

```
data2=pd.read_csv("서울특별시_강동구_CCTV_20180816.csv",sep=",",encoding='euc-kr')
```

```
data2.loc[data2['설치목적구분']=='생활방범',:]
```

```
newdata=data2.loc[data2['설치목적구분']=='생활방범',['위도','경도']]
newdata
```

```
newdata.to_csv('newdata.csv', index=False)
```

```
Gangdong_map=folium.Map(location=[37.52996,127.1235], zoom_start=13) #location은 강동구청

for n in newdata.index:
  folium.CircleMarker([newdata['위도'][n],newdata['경도'][n]], radius=8, color='#b23776',fill_color='#b23776').add_to(Gangdong_map)

Gangdong_map #강동구 방범용 CCTV위치
```
